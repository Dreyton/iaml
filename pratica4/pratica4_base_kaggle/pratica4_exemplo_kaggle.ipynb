```python
import pandas as pd

df = pd.read_csv("heart_attack_prediction_dataset.csv");
df
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df = df.drop(['Patient ID','Hemisphere', 'Continent'], axis = 1)
df.head()
```

```python
df['BP_systolic'] = df['Blood Pressure'].apply(lambda x: x.split("/")[0])
df['BP_diastolic'] = df['Blood Pressure'].apply(lambda x: x.split("/")[1])
df.head()
```

```python
df = df.drop(['Country','Diet'], axis = 1)
df.head()
```

```python
df = df.drop(['Sex','Blood Pressure'], axis = 1)
df.head()
```

```python
df.dtypes
```

```python
X = df.drop('Heart Attack Risk', axis = 1)
y = df['Heart Attack Risk']
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled,  y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(hidden_layer_sizes=(100), max_iter = 1000)
model.fit(X_train, y_train)
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=10)
model.fit(X_train, y_train)
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
