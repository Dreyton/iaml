```python
fig = plt.figure(figsize=(7, 3))
axes = fig.add_axes([0, 0, 1, 1])
f1 = 'sepal length (cm)'
f2 = 'sepal width (cm)'

c = {0: 'r', 1: 'b', 2: 'k'}

for specie in y.unique():
  axes.scatter(X[y==specie][f1], X[y==specie][f2], color=c[specie])
axes.set_xlabel('Sepal length (cm)')
axes.set_ylabel('Sepal width (cm)')
axes.legend(['Setosa', 'Versicolor', 'Virginica'])
```
